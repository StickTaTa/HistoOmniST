# HisToGene Patch-H5 Single-Slide Benchmark Smoke

This notebook records two engineering checks for the external HisToGene-style baseline path: truncated predictions are rejected before benchmark evaluation, and a complete single-slide prediction bundle can be evaluated by the coverage95 benchmark harness. These are smoke checks, not formal baseline performance results.

```python
import json
from pathlib import Path

ROOT = Path.cwd()
if not (ROOT / "src" / "histoomnist").exists() and (ROOT.parent / "src" / "histoomnist").exists():
    ROOT = ROOT.parent

refusal_summary = json.loads(
    (ROOT / "results/hest1k_human_visium_expression/external_baselines/histogene_patch_h5_refusal_smoke_predictions/prediction_summary.json")
    .read_text(encoding="utf-8")
)
complete_summary = json.loads(
    (ROOT / "results/hest1k_human_visium_expression/external_baselines/histogene_patch_h5_single_slide_smoke_predictions/prediction_summary.json")
    .read_text(encoding="utf-8")
)
benchmark_summary = json.loads(
    (ROOT / "results/hest1k_human_visium_expression/benchmark_results/histogene_patch_h5_single_slide_smoke/run_summary.json")
    .read_text(encoding="utf-8")
)
```

```python
refusal_slide = refusal_summary["slides"][0]
refusal_check = {
    "sample_id": refusal_slide["sample_id"],
    "n_predicted_spots": refusal_slide["n_predicted_spots"],
    "expected_spots": refusal_slide["expected_spots"],
    "complete_slide_prediction": refusal_slide["complete_slide_prediction"],
    "benchmark_evaluable_without_truncation": refusal_summary["benchmark_evaluable_without_truncation"],
}
assert refusal_check["n_predicted_spots"] == 4
assert refusal_check["expected_spots"] == 3385
assert refusal_check["complete_slide_prediction"] is False
assert refusal_check["benchmark_evaluable_without_truncation"] is False
refusal_check
```

```python
complete_slide = complete_summary["slides"][0]
benchmark_check = {
    "sample_id": complete_slide["sample_id"],
    "n_predicted_spots": complete_slide["n_predicted_spots"],
    "expected_spots": complete_slide["expected_spots"],
    "complete_slide_prediction": complete_slide["complete_slide_prediction"],
    "benchmark_evaluable_without_truncation": complete_summary["benchmark_evaluable_without_truncation"],
    "benchmark_n_slides": benchmark_summary["n_slides"],
    "n_common_genes": benchmark_summary["n_common_genes"],
    "valid_genes": benchmark_summary["gene_metrics"]["valid_genes"],
    "mean_gene_pearson": benchmark_summary["gene_metrics"]["mean_gene_pearson"],
    "median_gene_pearson": benchmark_summary["gene_metrics"]["median_gene_pearson"],
}
assert benchmark_check["n_predicted_spots"] == 3385
assert benchmark_check["expected_spots"] == 3385
assert benchmark_check["complete_slide_prediction"] is True
assert benchmark_check["benchmark_evaluable_without_truncation"] is True
assert benchmark_check["benchmark_n_slides"] == 1
assert benchmark_check["n_common_genes"] == 16942
assert benchmark_check["valid_genes"] == 16396
benchmark_check
```
